```python
import os
PKG = '/kaggle/working/nicto_pkg'
os.makedirs(f'{PKG}/nicto_neural/neural', exist_ok=True)
for p in [f'{PKG}/nicto_neural/__init__.py', f'{PKG}/nicto_neural/neural/__init__.py']:
    with open(p, 'w') as f: f.write('')
import sys; sys.path.insert(0, PKG)

# Write super_config.py
with open(f'{PKG}/nicto_neural/neural/super_config.py', 'w') as f:
    f.write("import torch\nfrom dataclasses import dataclass\nfrom typing import Optional\n\n@dataclass\nclass SuperConfig:\n    d_model: int = 512\n    n_heads: int = 8\n    n_kv_heads: Optional[int] = None\n    n_layers: int = 8\n    d_ff: int = 2048\n    n_experts: int = 4\n    n_active_experts: int = 1\n    expert_capacity_factor: float = 1.25\n    vocab_size: int = 8192\n    max_seq_len: int = 1024\n    dropout: float = 0.1\n    eps: float = 1e-6\n    use_rope: bool = True\n    rope_theta: float = 10000.0\n    rope_scaling: Optional[dict] = None\n    use_flash_attn: bool = False\n\n    def __post_init__(self):\n        if self.n_kv_heads is None:\n            self.n_kv_heads = self.n_heads\n\n    @property\n    def head_dim(self):\n        return self.d_model // self.n_heads\n\n    def estimate_params(self):\n        total = (self.vocab_size * self.d_model +\n                 4 * self.d_model * self.d_model * self.n_layers +\n                 self.n_experts * 3 * self.d_model * self.d_ff * self.n_layers)\n        return {'total': total, 'total_billions': total / 1e9}\n")

# Write super_core.py
with open(f'{PKG}/nicto_neural/neural/super_core.py', 'w') as f:
    f.write("import torch\nimport torch.nn as nn\nimport torch.nn.functional as F\nimport math\nfrom .super_config import SuperConfig\n\nclass RMSNorm(nn.Module):\n    def __init__(self, dim, eps=1e-6):\n        super().__init__()\n        self.eps = eps\n        self.weight = nn.Parameter(torch.ones(dim))\n\n    def forward(self, x):\n        return x * torch.rsqrt(x.pow(2).mean(-1, keepdim=True) + self.eps) * self.weight\n\ndef precompute_freqs(dim, max_len, theta=10000.0):\n    freqs = 1.0 / (theta ** (torch.arange(0, dim, 2).float() / dim))\n    t = torch.arange(max_len).float()\n    angles = torch.outer(t, freqs)\n    return torch.cos(angles), torch.sin(angles)\n\nclass Attention(nn.Module):\n    def __init__(self, cfg):\n        super().__init__()\n        D, H, KV = cfg.d_model, cfg.n_heads, cfg.n_kv_heads\n        self.hd = cfg.head_dim\n        self.H = H\n        self.n_rep = H // KV\n        self.q = nn.Linear(D, H * self.hd, bias=False)\n        self.k = nn.Linear(D, KV * self.hd, bias=False)\n        self.v = nn.Linear(D, KV * self.hd, bias=False)\n        self.o = nn.Linear(H * self.hd, D, bias=False)\n\n    def forward(self, x, cos=None, sin=None):\n        B, T, _ = x.shape\n        q = self.q(x).view(B, T, self.H, self.hd)\n        k = self.k(x).view(B, T, self.H // self.n_rep, self.hd)\n        v = self.v(x).view(B, T, self.H // self.n_rep, self.hd)\n        if self.n_rep > 1:\n            k = k[:, :, :, None, :].expand(B, T, self.H // self.n_rep, self.n_rep, self.hd).reshape(B, T, self.H, self.hd)\n            v = v[:, :, :, None, :].expand(B, T, self.H // self.n_rep, self.n_rep, self.hd).reshape(B, T, self.H, self.hd)\n        if cos is not None:\n            hh = self.hd // 2\n            q1, q2 = q[..., :hh], q[..., hh:]\n            k1, k2 = k[..., :hh], k[..., hh:]\n            cos_t = cos[:T].unsqueeze(0).unsqueeze(0)\n            sin_t = sin[:T].unsqueeze(0).unsqueeze(0)\n            q = torch.cat([q1 * cos_t - q2 * sin_t, q1 * sin_t + q2 * cos_t], -1)\n            k = torch.cat([k1 * cos_t - k2 * sin_t, k1 * sin_t + k2 * cos_t], -1)\n        q, k, v = q.transpose(1, 2), k.transpose(1, 2), v.transpose(1, 2)\n        attn = (q @ k.transpose(-2, -1)) * (self.hd ** -0.5)\n        mask = torch.triu(torch.full((T, T), float('-inf'), device=x.device), diagonal=1)\n        attn = attn + mask\n        attn = F.softmax(attn, dim=-1, dtype=torch.float32).to(x.dtype)\n        return self.o((attn @ v).transpose(1, 2).contiguous().view(B, T, -1))\n\nclass SwiGLU(nn.Module):\n    def __init__(self, D, FF):\n        super().__init__()\n        self.w1 = nn.Linear(D, FF, bias=False)\n        self.w2 = nn.Linear(FF, D, bias=False)\n        self.w3 = nn.Linear(D, FF, bias=False)\n\n    def forward(self, x):\n        return self.w2(F.silu(self.w1(x)) * self.w3(x))\n\nclass MoE(nn.Module):\n    def __init__(self, cfg):\n        super().__init__()\n        self.router = nn.Linear(cfg.d_model, cfg.n_experts, bias=False)\n        self.experts = nn.ModuleList([SwiGLU(cfg.d_model, cfg.d_ff) for _ in range(cfg.n_experts)])\n        self.k = cfg.n_active_experts\n\n    def forward(self, x):\n        B, T, D = x.shape\n        flat = x.view(-1, D)\n        w, idx = torch.topk(F.softmax(self.router(flat), -1), self.k, -1)\n        w = w / w.sum(-1, keepdim=True)\n        out = torch.zeros_like(flat)\n        for e in range(len(self.experts)):\n            mask = (idx == e).any(-1)\n            if mask.any():\n                ids = mask.nonzero(as_tuple=True)[0]\n                ew = w[ids] * (idx[ids] == e).float()\n                out[ids] += self.experts[e](flat[ids]) * ew.sum(-1, keepdim=True)\n        return out.view(B, T, D)\n\nclass Block(nn.Module):\n    def __init__(self, cfg):\n        super().__init__()\n        self.an = RMSNorm(cfg.d_model)\n        self.attn = Attention(cfg)\n        self.mn = RMSNorm(cfg.d_model)\n        self.moe = MoE(cfg)\n\n    def forward(self, x, cos=None, sin=None):\n        return x + self.moe(self.mn(x + self.attn(self.an(x), cos, sin)))\n\nclass SuperNeuralCore(nn.Module):\n    def __init__(self, cfg):\n        super().__init__()\n        self.cfg = cfg\n        self.wte = nn.Embedding(cfg.vocab_size, cfg.d_model)\n        self.cos, self.sin = precompute_freqs(cfg.head_dim, cfg.max_seq_len, cfg.rope_theta)\n        self.layers = nn.ModuleList([Block(cfg) for _ in range(cfg.n_layers)])\n        self.norm = RMSNorm(cfg.d_model)\n        self.lm = nn.Linear(cfg.d_model, cfg.vocab_size, bias=False)\n        self.lm.weight = self.wte.weight\n        self.apply(lambda m: nn.init.normal_(m.weight, 0, 0.02) if isinstance(m, (nn.Linear, nn.Embedding)) else None)\n\n    def forward(self, ids):\n        x = self.wte(ids)\n        c = self.cos[:ids.shape[1]].to(ids.device)\n        s = self.sin[:ids.shape[1]].to(ids.device)\n        for layer in self.layers:\n            x = layer(x, c, s)\n        return {'logits': self.lm(self.norm(x)), 'hidden': x}\n")

print('Source files written')
```

## GENERATE TRAINING DATA

```python
# Code-related Q&A pairs for training
QAS = [
    ("Reverse a string", "def rev(s): return s[::-1]"),
    ("Binary search", "def bs(a,t):\n l,r=0,len(a)-1\n while l<=r:\n  m=(l+r)//2\n  if a[m]==t: return m\n  if a[m]<t: l=m+1\n  else: r=m-1\n  return -1"),
    ("What is a linked list?", "Each node stores data + pointer. O(1) insert/delete."),
    ("Stack class", "class Stack:\n def __init__(s): s.i=[]\n def push(s,x): s.i.append(x)\n def pop(s): return s.i.pop() if s.i else None"),
    ("Prime check", "def is_prime(n):\n if n<2: return False\n for i in range(2,int(n**0.5)+1):\n  if n%i==0: return False\n  return True"),
    ("Hash table", "Key-value store. O(1) avg lookups."),
    ("Factorial", "def fact(n): return 1 if n<=1 else n*fact(n-1)"),
    ("Quicksort", "def qs(a):\n if len(a)<=1: return a\n p=a[len(a)//2]\n return qs([x for x in a if x<p])+[x for x in a if x==p]+qs([x for x in a if x>p])"),
    ("SQL injection", "Use parameterized queries."),
    ("Flask API", "from flask import Flask,jsonify,request\napp=Flask(__name__)"),
    ("What is OOP?", "Object-Oriented Programming: classes, objects, inheritance, polymorphism, encapsulation."),
    ("Bubble sort", "def bs(a):\n for i in range(len(a)):\n  for j in range(len(a)-1):\n   if a[j]>a[j+1]: a[j],a[j+1]=a[j+1],a[j]\n  return a"),
    ("Regex for email", "import re\npat=r'\\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Z|a-z]{2,}\\b'"),
    ("Merge sort", "def ms(a):\n if len(a)<=1: return a\n m=len(a)//2\n return merge(ms(a[:m]),ms(a[m:]))\ndef merge(l,r):\n i=j=0; res=[]\n while i<len(l) and j<len(r):\n  if l[i]<r[j]: res.append(l[i]); i+=1\n  else: res.append(r[j]); j+=1\n  return res+l[i:]+r[j:]"),
    ("HTTP status codes", "200=OK, 201=Created, 400=Bad Request, 401=Unauthorized, 403=Forbidden, 404=Not Found, 500=Internal Server Error"),
]
all_c = []
for q, a in QAS:
    for _ in range(200):
        all_c.append(f'<|user|>{q}<|/|>\n<|assistant|>{a}<|/|>')
print(f'Generated {len(all_c)} conversations')
```

## TRAIN TOKENIZER

```python
from tokenizers import Tokenizer, models, trainers, pre_tokenizers, decoders, processors

corpus = '/kaggle/working/corpus.txt'
with open(corpus, 'w') as f:
    for c in all_c:
        f.write(c + '\n')

tok = Tokenizer(models.BPE())
tok.pre_tokenizer = pre_tokenizers.ByteLevel(add_prefix_space=False)
tok.decoder = decoders.ByteLevel()
tok.post_processor = processors.ByteLevel(trim_offsets=True)

SPECIAL = ['<|pad|>', '<|bos|>', '<|eos|>', '<|user|>', '<|assistant|>', '<|code_start|>', '<|code_end|>']
trainer = trainers.BpeTrainer(vocab_size=4096, special_tokens=SPECIAL, min_frequency=1, show_progress=True)
tok.train([corpus], trainer)
tok.save('/kaggle/working/tok.json')

VOCAB = tok.get_vocab_size()
PAD = tok.token_to_id('<|pad|>')
EOS = tok.token_to_id('<|eos|>')
print(f'Vocab: {VOCAB}, PAD={PAD}, EOS={EOS}')
```

## INIT MODEL (~100M)

```python
cfg = SuperConfig(d_model=512, n_heads=8, n_kv_heads=4, n_layers=8, d_ff=2048,
                  n_experts=4, n_active_experts=1, max_seq_len=1024,
                  vocab_size=VOCAB, dropout=0.1, use_rope=True, rope_theta=10000.0)
est = cfg.estimate_params()
print(f'd={cfg.d_model}, L={cfg.n_layers}, H={cfg.n_heads}, E={cfg.n_experts}')
print(f'Params: {est["total_billions"]:.3f}B ({est["total"]:,})')

model = SuperNeuralCore(cfg).to(DEVICE)
total = sum(p.numel() for p in model.parameters())
print(f'Actual: {total:,} ({total/1e9:.3f}B)')
print(f'Device: {DEVICE}')
```

## TOKENIZE + SPLIT

```python
import random, numpy as np

def enc(t):
    return tok.encode(t).ids[:cfg.max_seq_len]

ids = [enc(c) for c in all_c if len(enc(c)) >= 5]
print(f'{len(ids)} valid seqs, avg {np.mean([len(t) for t in ids]):.1f} tok')

random.shuffle(ids)
split = int(len(ids) * 0.95)
tr_ids, val_ids = ids[:split], ids[split:]
print(f'Train: {len(tr_ids)}, Val: {len(val_ids)}')
```

## TRAINING LOOP
SFT next-token prediction with cosine LR.

```python
BS = 12
GA = 4  # gradient accumulation steps
LR = 3e-4
EPOCHS = 3

opt = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=0.1)

def collate(ts):
    ml = min(max(len(t) for t in ts), cfg.max_seq_len)
    ids_t = torch.full((len(ts), ml), PAD, dtype=torch.long, device=DEVICE)
    lbls = torch.full((len(ts), ml), -100, dtype=torch.long, device=DEVICE)
    for i, t in enumerate(ts):
        sl = min(len(t), ml)
        ids_t[i, :sl] = torch.tensor(t[:sl], dtype=torch.long)
        lbls[i, :sl] = torch.tensor(t[:sl], dtype=torch.long)
    return ids_t, lbls

def batches(data, bs):
    ix = list(range(len(data)))
    random.shuffle(ix)
    for i in range(0, len(ix), bs):
        yield collate([data[j] for j in ix[i:i+bs]])

steps_per_epoch = len(tr_ids) // (BS * GA)
total_steps = steps_per_epoch * EPOCHS
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(opt, total_steps, 1e-5)
print(f'Steps/epoch: {steps_per_epoch}, Total: {total_steps}')
```

```python
log = []
global_step = 0
best_val = float('inf')

for ep in range(1, EPOCHS + 1):
    print(f'\n===== EPOCH {ep}/{EPOCHS} =====')
    model.train()
    losses = []
    opt.zero_grad()

    for bidx, (inp, lbl) in enumerate(batches(tr_ids, BS)):
        out = model(inp)
        logits = out['logits'][..., :-1, :].contiguous()
        targets = lbl[..., 1:].contiguous()
        loss = F.cross_entropy(logits.view(-1, VOCAB), targets.view(-1), ignore_index=-100)
        (loss / GA).backward()

        if (bidx + 1) % GA == 0:
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()
            scheduler.step()
            opt.zero_grad()
            global_step += 1

            with torch.no_grad():
                acc = (logits.argmax(-1) == targets).float().mean().item()
            losses.append(loss.item())

            if global_step % 10 == 0:
                avg = np.mean(losses[-50:]) if len(losses) >= 50 else np.mean(losses)
                print(f'  Step {global_step:>4}/{total_steps} | loss:{loss.item():.4f} avg:{avg:.4f} acc:{acc:.4f}')

            log.append({'step': global_step, 'loss': loss.item(), 'acc': acc})

    train_loss = np.mean(losses)
    print(f'Train loss: {train_loss:.4f}')

    # validation
    model.eval()
    val_losses = []
    with torch.no_grad():
        for inp, lbl in batches(val_ids, BS):
            v = F.cross_entropy(
                model(inp)['logits'][..., :-1, :].contiguous().view(-1, VOCAB),
                lbl[..., 1:].contiguous().view(-1),
                ignore_index=-100
            )
            val_losses.append(v.item())

    val_loss = np.mean(val_losses)
    val_ppl = math.exp(min(val_loss, 20))
    print(f'Val: loss={val_loss:.4f} ppl={val_ppl:.2f}')

    # save checkpoint
    os.makedirs(f'/kaggle/working/ckpt/ep{ep}', exist_ok=True)
    torch.save({
        'model': model.state_dict(),
        'epoch': ep,
        'val_loss': val_loss
    }, f'/kaggle/working/ckpt/ep{ep}/model.pt')

    if val_loss < best_val:
        best_val = val_loss
        torch.save({'model': model.state_dict(), 'cfg': cfg}, '/kaggle/working/best.pt')
        print(f'*** NEW BEST! val_loss={val_loss:.4f} ***')

    gc.collect()
    torch.cuda.empty_cache()

print(f'\nTraining complete! Steps: {global_step}, Best val: {best_val:.4f}')
```

## RLCF (Reinforcement Learning from Compiler Feedback)

```python
def reward(code):
    try:
        ast.parse(code)
        return 1.0
    except SyntaxError:
        return -0.5

@torch.no_grad()
def generate(prompt_ids, max_new=50, temperature=0.8):
    model.eval()
    inp = torch.tensor([prompt_ids], dtype=torch.long, device=DEVICE)
    for _ in range(max_new):
        logits = model(inp)['logits'][:, -1, :] / temperature
        # top-40 sampling
        vals, _ = torch.topk(logits, 40, -1)
        logits[logits < vals[:, -1:]] = float('-inf')
        next_id = torch.multinomial(F.softmax(logits, -1), 1)
        inp = torch.cat([inp, next_id], -1)
        if next_id.item() == EOS:
            break
    return inp[0].tolist()

# Prepare code-start token
code_start_id = tok.encode('<|code_start|>').ids[0] if tok.encode('<|code_start|>').ids else VOCAB - 1

rl_opt = torch.optim.AdamW(model.parameters(), lr=1e-5, weight_decay=0.01)
N_RLCF_STEPS = 30
rl_log = []

for step in range(N_RLCF_STEPS):
    model.train()
    # pick random training example, append code_start
    ctx = random.choice(tr_ids)[:random.randint(20, 50)] + [code_start_id]
    gen_ids = generate(ctx, 40)
    gen_text = tok.decode(gen_ids[len(ctx):])
    # extract code before special tokens
    for sep in ['<|code_end|>', '<|eos|>']:
        if sep in gen_text:
            gen_text = gen_text.split(sep)[0]
    r = reward(gen_text)

    inp = torch.tensor([gen_ids], dtype=torch.long, device=DEVICE)
    out = model(inp)
    shift_logits = out['logits'][0, :-1]
    shift_targets = inp[0, 1:]
    ce_loss = F.cross_entropy(shift_logits.view(-1, VOCAB), shift_targets.view(-1), ignore_index=PAD)

    # Penalize on negative reward, reinforce on positive
    weight = max(0, 1.0 - r)
    (ce_loss * weight).backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
    rl_opt.step()
    rl_opt.zero_grad()

    rl_log.append({'reward': r, 'loss': ce_loss.item()})

    if step % 10 == 0:
        avg_r = np.mean([x['reward'] for x in rl_log[-10:]])
        print(f'RLCF {step}/{N_RLCF_STEPS} | r={r:.2f} | avg_r={avg_r:.2f} | ce={ce_loss.item():.4f}')

avg_rl_reward = np.mean([x['reward'] for x in rl_log[-10:]])
print(f'\nRLCF done! Avg reward (last 10): {avg_rl_reward:.3f}')
```

```python
prompts = [
    '<|user|>Reverse a string in Python<|/|>\n<|assistant|>\n<|code_start|>',
    '<|user|>What is a hash table?<|/|>\n<|assistant|>',
    '<|user|>Write quicksort<|/|>\n<|assistant|>\n<|code_start|>',
]
print('=' * 60)
print('VALIDATION GENERATIONS')
print('=' * 60)
for prompt in prompts:
    ids = tok.encode(prompt).ids
    gen_ids = generate(ids, 80, 0.7)
    text = tok.decode(gen_ids)
    print(f'\nPrompt: {prompt[:60]}...')
    print(f'Output: {text[-300:]}')
    m = re.search(r'<\|code_start\|>(.*?)(?:<\|code_end\|>|<\|eos\|>|$)', text, re.DOTALL)
    if m:
        print(f'Code reward: {reward(m.group(1)):.2f}')
    print('-' * 40)
```
