```python
import json
data_path = f"{DATA_DIR}/super_v3_chatml.jsonl"
entries = []
with open(data_path, 'r') as f:
    for line in f:
        entries.append(json.loads(line))
print(f"Loaded {len(entries)} training entries")

# Analyze categories
from collections import Counter
cats = Counter()
for e in entries:
    m = e.get("metadata", {})
    if isinstance(m, dict):
        cats[m.get("domain", m.get("category", "unknown"))] += 1
print(f"Domains: {len(cats)}")
for dom, cnt in cats.most_common(10):
    print(f"  {dom}: {cnt}")
```

## 3. Training Configuration
4 base models ├ù 5 adapters = 20 trainings

```python
BASE_MODELS = {
    "phi3": {"model": "microsoft/Phi-3-mini-4k-instruct", "params": "3.8B", "seq_len": 4096},
    "llama32": {"model": "meta-llama/Llama-3.2-3B-Instruct", "params": "3B", "seq_len": 8192},
    "qwen25": {"model": "Qwen/Qwen2.5-7B-Instruct", "params": "7B", "seq_len": 32768},
    "mistral": {"model": "mistralai/Mistral-7B-Instruct-v0.3", "params": "7B", "seq_len": 32768},
}

ADAPTERS = [
    {"name": "universal", "desc": "General purpose ΓÇö all 361K entries"},
    {"name": "kyros", "desc": "Lightweight AI ΓÇö 55K entries"},
    {"name": "omega", "desc": "Balanced reasoning ΓÇö 150K entries"},
    {"name": "main", "desc": "Full-featured with security ΓÇö 119K entries"},
    {"name": "x", "desc": "Frontier agent orchestrator ΓÇö 125K entries"},
]

LORA_RANK = 16
LORA_ALPHA = 16
BATCH_SIZE = 2 if torch.cuda.get_device_properties(0).total_memory / 1e9 < 20 else 4
GRAD_ACCUM = 4
EPOCHS = 2
print(f"Batch size: {BATCH_SIZE}, Grad accum: {GRAD_ACCUM}, Epochs: {EPOCHS}")
print(f"Total trainings: {len(BASE_MODELS)} models ├ù {len(ADAPTERS)} adapters = {len(BASE_MODELS) * len(ADAPTERS)}")
```

## 4. Unsloth Training Engine

```python
from unsloth import FastLanguageModel, is_bfloat16_supported
from unsloth.chat_templates import get_chat_template
from transformers import TrainingArguments
from trl import SFTTrainer, DataCollatorForCompletionOnlyLM
import torch

results_log = []

def train_adapter(base_key: str, adapter_name: str, num_entries: int):
    """Train one LoRA adapter on a base model."""
    model_cfg = BASE_MODELS[base_key]
    model_id = model_cfg["model"]
    out_dir = f"/kaggle/working/models/{base_key}/{adapter_name}"
    os.makedirs(out_dir, exist_ok=True)
    
    print(f"\n{'='*60}")
    print(f"Training {base_key}/{adapter_name} on {num_entries} entries")
    print(f"{'='*60}")
    
    # Load base model in 4-bit
    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name=model_id,
        max_seq_length=model_cfg["seq_len"],
        dtype=None,
        load_in_4bit=True,
    )
    
    # Apply LoRA
    model = FastLanguageModel.get_peft_model(
        model,
        r=LORA_RANK,
        lora_alpha=LORA_ALPHA,
        target_modules=["q_proj", "k_proj", "v_proj", "o_proj",
                       "gate_proj", "up_proj", "down_proj"],
        lora_dropout=0,
        bias="none",
        use_gradient_checkpointing="unsloth",
        random_state=42,
        use_rslora=True,
    )
    
    # ChatML tokenizer
    tokenizer = get_chat_template(tokenizer, chat_template="chatml")
    
    # Format training data
    def format_entries(limit):
        texts = []
        for e in entries[:limit]:
            msgs = e.get("messages", [])
            parts = []
            for m in msgs:
                role = m.get("role", "user")
                content = m.get("content", "")
                parts.append(f"<|im_start|>{role}\n{content}<|im_end|>")
            text = "\n".join(parts) + "\n<|im_start|>assistant\n"
            texts.append(text)
        return texts[:limit]
    
    train_texts = format_entries(num_entries)
    print(f"  Formatted {len(train_texts)} training texts")
    
    # Training args
    training_args = TrainingArguments(
        output_dir=out_dir,
        per_device_train_batch_size=BATCH_SIZE,
        gradient_accumulation_steps=GRAD_ACCUM,
        warmup_steps=20,
        learning_rate=2e-4,
        fp16=not is_bfloat16_supported(),
        bf16=is_bfloat16_supported(),
        logging_steps=10,
        optim="adamw_8bit",
        weight_decay=0.01,
        lr_scheduler_type="cosine",
        seed=42,
        max_grad_norm=0.3,
        num_train_epochs=EPOCHS,
        report_to="none",
    )
    
    # Trainer
    trainer = SFTTrainer(
        model=model,
        tokenizer=tokenizer,
        train_dataset=train_texts,
        max_seq_length=model_cfg["seq_len"],
        dataset_num_proc=2,
        packing=False,
        args=training_args,
    )
    
    # Train
    t0 = time.time()
    trainer_stats = trainer.train()
    train_time = time.time() - t0
    final_loss = trainer_stats.training_loss if hasattr(trainer_stats, 'training_loss') else trainer_stats.metrics.get('train_loss', 0)
    
    # Save LoRA
    model.save_pretrained(out_dir)
    tokenizer.save_pretrained(out_dir)
    
    # Export GGUF (q4_k_m)
    gguf_path = f"{out_dir}/model.q4_k_m.gguf"
    model.save_pretrained_gguf(gguf_path, quantization_method="q4_k_m")
    
    # Clean up
    del model, tokenizer, trainer
    gc.collect()
    torch.cuda.empty_cache()
    
    result = {
        "base_model": base_key,
        "adapter": adapter_name,
        "model_id": model_id,
        "entries": num_entries,
        "train_time_min": round(train_time / 60, 1),
        "final_loss": round(final_loss, 4) if isinstance(final_loss, float) else float(final_loss),
        "gguf_path": gguf_path,
    }
    results_log.append(result)
    print(f"  Done: {train_time/60:.1f}min, loss={final_loss:.4f}")
    return result
```

## 5. Run All Trainings

```python
ADAPTER_CONFIGS = {
    "universal": 360000,
    "kyros": 55000,
    "omega": 150000,
    "main": 119000,
    "x": 125000,
}

overall_start = time.time()
completed = 0

for base_key in BASE_MODELS:
    print(f"\n\n{'#'*70}")
    print(f"# TRAINING BASE MODEL: {base_key} ({BASE_MODELS[base_key]['params']})")
    print(f"{'#'*70}")
    for adapter in ADAPTERS:
        name = adapter["name"]
        try:
            train_adapter(base_key, name, ADAPTER_CONFIGS[name])
            completed += 1
        except Exception as e:
            print(f"  FAILED {base_key}/{name}: {e}")
            results_log.append({"base_model": base_key, "adapter": name, "status": "failed", "error": str(e)[:100]})
        gc.collect()
        torch.cuda.empty_cache()

total_time = time.time() - overall_start
print(f"\n\n{'='*60}")
print(f"ALL DONE: {completed}/{len(BASE_MODELS) * len(ADAPTERS)} trainings completed")
print(f"Total time: {total_time/3600:.1f} hours")
print(f"{'='*60}")
```

## 6. Results Summary

```python
report = {
    "training_completed": completed,
    "total_time_hours": round(total_time / 3600, 1),
    "results": results_log,
}
with open("/kaggle/working/training_report.json", "w") as f:
    json.dump(report, f, indent=2)

print(json.dumps(report, indent=2))
print(f"\nModels saved in: /kaggle/working/models/")
print(f"Report saved: /kaggle/working/training_report.json")
!ls -lhR /kaggle/working/models/ 2>/dev/null | grep ".gguf"
```
