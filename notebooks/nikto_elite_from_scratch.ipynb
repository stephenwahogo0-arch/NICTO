```python
SC_SRC = '''
import torch
from dataclasses import dataclass
from typing import Optional

@dataclass
class SuperConfig:
    d_model: int = 768
    n_heads: int = 12
    n_kv_heads: Optional[int] = None
    n_layers: int = 12
    d_ff: int = 2048
    n_experts: int = 4
    n_active_experts: int = 1
    expert_capacity_factor: float = 1.25
    vocab_size: int = 32768
    max_seq_len: int = 2048
    dropout: float = 0.1
    eps: float = 1e-6
    n_brain_heads: int = 9
    brain_head_dim: int = 0
    reasoning_paths: int = 6
    reasoning_dim: int = 0
    use_rope: bool = True
    rope_theta: float = 10000.0
    rope_scaling: Optional[dict] = None
    use_flash_attn: bool = True

    def __post_init__(self):
        if self.n_kv_heads is None:
            self.n_kv_heads = self.n_heads
        if self.brain_head_dim == 0:
            self.brain_head_dim = self.d_model // max(1, self.n_brain_heads)
        if self.reasoning_dim == 0:
            self.reasoning_dim = self.d_model

    @property
    def head_dim(self) -> int:
        return self.d_model // self.n_heads

    def estimate_params(self) -> dict:
        emb = self.vocab_size * self.d_model
        attn = 4 * self.d_model * self.d_model * self.n_layers
        moe = self.n_experts * 3 * self.d_model * self.d_ff * self.n_layers
        total = emb + attn + moe
        return {"embedding": emb, "attention": attn, "moe": moe,
                "total": total, "total_billions": total / 1e9}
'''
with open(f'{PKG}/nicto_neural/neural/super_config.py', 'w') as f: f.write(SC_SRC)
print('super_config.py written')
```

```python
SC_CORE_SRC = '''
import torch
import torch.nn as nn
import torch.nn.functional as F
import math
from typing import Optional, Tuple, List, Dict
from .super_config import SuperConfig

class RMSNorm(nn.Module):
    def __init__(self, dim, eps=1e-6):
        super().__init__()
        self.eps = eps
        self.weight = nn.Parameter(torch.ones(dim))
    def forward(self, x):
        return x * torch.rsqrt(x.pow(2).mean(-1, keepdim=True) + self.eps) * self.weight

def precompute_freqs(dim, max_len, theta=10000.0):
    freqs = 1.0 / (theta ** (torch.arange(0, dim, 2).float() / dim))
    t = torch.arange(max_len).float()
    angles = torch.outer(t, freqs)
    return torch.cos(angles), torch.sin(angles)

def apply_rope(x, cos, sin):
    half = x.shape[-1] // 2
    x1, x2 = x[..., :half], x[..., half:]
    cos = cos[:x.shape[-2], :].unsqueeze(0).unsqueeze(0)
    sin = sin[:x.shape[-2], :].unsqueeze(0).unsqueeze(0)
    return torch.cat([x1 * cos - x2 * sin, x1 * sin + x2 * cos], -1)

class FlashAttention(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.d_model = config.d_model
        self.n_heads = config.n_heads
        self.n_kv = config.n_kv_heads
        self.head_dim = config.head_dim
        self.n_rep = self.n_heads // self.n_kv
        self.q = nn.Linear(config.d_model, self.n_heads * self.head_dim, bias=False)
        self.k = nn.Linear(config.d_model, self.n_kv * self.head_dim, bias=False)
        self.v = nn.Linear(config.d_model, self.n_kv * self.head_dim, bias=False)
        self.o = nn.Linear(self.n_heads * self.head_dim, config.d_model, bias=False)
        self.scale = self.head_dim ** -0.5
        self.drop = nn.Dropout(config.dropout)

    def forward(self, x, mask=None, cos=None, sin=None):
        B, T, _ = x.shape
        q = self.q(x).view(B, T, self.n_heads, self.head_dim)
        k = self.k(x).view(B, T, self.n_kv, self.head_dim)
        v = self.v(x).view(B, T, self.n_kv, self.head_dim)
        if cos is not None:
            q, k = apply_rope(q, cos, sin), apply_rope(k, cos, sin)
        if self.n_rep > 1:
            k = k[:, :, :, None, :].expand(B, T, self.n_kv, self.n_rep, self.head_dim).reshape(B, T, self.n_heads, self.head_dim)
            v = v[:, :, :, None, :].expand(B, T, self.n_kv, self.n_rep, self.head_dim).reshape(B, T, self.n_heads, self.head_dim)
        q, k, v = q.transpose(1, 2), k.transpose(1, 2), v.transpose(1, 2)
        attn = torch.matmul(q, k.transpose(-2, -1)) * self.scale
        if mask is not None:
            attn = attn + mask[:, :, :T, :T]
        attn = F.softmax(attn, dim=-1, dtype=torch.float32).to(x.dtype)
        attn = self.drop(attn)
        out = torch.matmul(attn, v).transpose(1, 2).contiguous().view(B, T, -1)
        return self.o(out)

class SwiGLU(nn.Module):
    def __init__(self, d_model, d_ff):
        super().__init__()
        self.w1 = nn.Linear(d_model, d_ff, bias=False)
        self.w2 = nn.Linear(d_ff, d_model, bias=False)
        self.w3 = nn.Linear(d_model, d_ff, bias=False)
    def forward(self, x):
        return self.w2(F.silu(self.w1(x)) * self.w3(x))

class MoE(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.n_experts = config.n_experts
        self.n_active = config.n_active_experts
        self.router = nn.Linear(config.d_model, config.n_experts, bias=False)
        self.experts = nn.ModuleList([SwiGLU(config.d_model, config.d_ff) for _ in range(config.n_experts)])
        self.aux_loss = 0.0
    def forward(self, x):
        B, T, D = x.shape
        flat = x.view(-1, D)
        logits = self.router(flat)
        weights, indices = torch.topk(F.softmax(logits, dim=-1), self.n_active, dim=-1)
        weights = weights / weights.sum(-1, keepdim=True)
        out = torch.zeros_like(flat)
        expert_counts = torch.zeros(self.n_experts, device=x.device)
        for e in range(self.n_experts):
            mask = (indices == e).any(-1)
            if mask.any():
                idx = mask.nonzero(as_tuple=True)[0]
                expert_out = self.experts[e](flat[idx])
                batch_weights = weights[idx]
                batch_indices = indices[idx]
                w = torch.where(batch_indices == e, batch_weights, torch.zeros_like(batch_weights))
                w = w.sum(-1, keepdim=True)
                out[idx] += expert_out * w
                expert_counts[e] = idx.numel()
        self.aux_loss = 0.01 * (expert_counts.var() / (expert_counts.mean() + 1e-8))
        return out.view(B, T, D)

class TransformerBlock(nn.Module):
    def __init__(self, config, layer_id):
        super().__init__()
        self.attn_norm = RMSNorm(config.d_model, config.eps)
        self.attn = FlashAttention(config)
        self.moe_norm = RMSNorm(config.d_model, config.eps)
        self.moe = MoE(config)
    def forward(self, x, mask=None, cos=None, sin=None):
        x = x + self.attn(self.attn_norm(x), mask=mask, cos=cos, sin=sin)
        x = x + self.moe(self.moe_norm(x))
        return x

class SuperNeuralCore(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.config = config
        self.wte = nn.Embedding(config.vocab_size, config.d_model)
        self.drop = nn.Dropout(config.dropout)
        self.cos, self.sin = precompute_freqs(config.head_dim, config.max_seq_len, config.rope_theta)
        self.layers = nn.ModuleList([TransformerBlock(config, i) for i in range(config.n_layers)])
        self.norm = RMSNorm(config.d_model, config.eps)
        self.lm_head = nn.Linear(config.d_model, config.vocab_size, bias=False)
        self.lm_head.weight = self.wte.weight
        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, (nn.Linear, nn.Embedding)):
                nn.init.normal_(m.weight, 0.0, 0.02)

    def forward(self, input_ids, mask=None):
        B, T = input_ids.shape
        x = self.drop(self.wte(input_ids))
        cos, sin = self.cos[:T].to(x.device), self.sin[:T].to(x.device)
        for layer in self.layers:
            x = layer(x, mask=mask, cos=cos, sin=sin)
        x = self.norm(x)
        return {"logits": self.lm_head(x), "hidden_states": x}
'''
with open(f'{PKG}/nicto_neural/neural/super_core.py', 'w') as f: f.write(SC_CORE_SRC)
print('super_core.py written')
from nicto_neural.neural.super_config import SuperConfig
from nicto_neural.neural.super_core import SuperNeuralCore
print('SuperNeuralCore imported successfully')
```

## PHASE 1: GENERATE TRAINING DATA
Self-contained: creates code conversations from built-in patterns.

```python
def gen_code_convos():
    patterns = [
        ('Write a Python function to reverse a string', 'def reverse_string(s):\n    return s[::-1]'),
        ('Implement binary search', 'def binary_search(arr, target):\n    l, r = 0, len(arr)-1\n    while l <= r:\n        mid = (l+r)//2\n        if arr[mid] == target: return mid\n        elif arr[mid] < target: l = mid+1\n        else: r = mid-1\n    return -1'),
        ('What is a linked list?', 'A linked list is a linear data structure where elements are stored in nodes. Each node contains data and a pointer to the next node. It allows O(1) insertions/deletions.'),
        ('Write a stack class', 'class Stack:\n    def __init__(self): self.items = []\n    def push(self, item): self.items.append(item)\n    def pop(self): return self.items.pop() if self.items else None\n    def is_empty(self): return len(self.items) == 0'),
        ('Check if a number is prime', 'def is_prime(n):\n    if n < 2: return False\n    for i in range(2, int(n**0.5)+1):\n        if n % i == 0: return False\n    return True'),
        ('What is a hash table?', 'A hash table maps keys to values using a hash function. Average O(1) for lookups, insertions, deletions. Collisions handled via chaining or open addressing.'),
        ('Write factorial recursively', 'def factorial(n): return 1 if n <= 1 else n * factorial(n-1)'),
        ('Write quicksort', 'def quicksort(arr):\n    if len(arr) <= 1: return arr\n    p = arr[len(arr)//2]\n    return quicksort([x for x in arr if x < p]) + [x for x in arr if x == p] + quicksort([x for x in arr if x > p])'),
        ('Explain SQL injection', 'SQL injection inserts malicious SQL into queries. Prevent with parameterized queries, input validation, least privilege.'),
        ('Find longest common prefix', 'def lcp(strs):\n    if not strs: return ""\n    p = strs[0]\n    for s in strs[1:]:\n        while s[:len(p)] != p and p: p = p[:-1]\n    return p'),
        ('Create a Flask API', 'from flask import Flask, jsonify, request\napp = Flask(__name__)\nn = []\n@app.route("/items")\ndef get(): return jsonify(n)\n@app.route("/items", methods=["POST"])\ndef post(): n.append(request.json); return jsonify(request.json), 201'),
        ('What is OOP?', 'Object-Oriented Programming organizes code into objects with data and methods. Pillars: encapsulation, inheritance, polymorphism, abstraction.'),
        ('Explain recursion', 'Recursion solves problems by having a function call itself. Requires: base case to stop, recursive case making progress toward base.'),
    ]
    convos = []
    for q, a in patterns:
        convos.append({'user': q, 'assistant': a})
    return convos

all_convos = gen_code_convos()
all_conversations = []
all_code = []
for c in all_convos:
    all_conversations.append(f'<|user|>' + c['user'] + '<|/|>\\n<|assistant|>' + c['assistant'] + '<|/|>')
    if 'def ' in c['assistant'] or 'class ' in c['assistant'] or 'from ' in c['assistant']:
        all_code.append(c['assistant'])
# Augment by repeating 100x
orig = list(all_conversations)
for _ in range(99):
    all_conversations.extend(orig)
print(f'Generated {len(all_conversations)} conversations')
print(f'Code blocks: {len(all_code)}')
```

## PHASE 2: CUSTOM CODE TOKENIZER

```python
from tokenizers import Tokenizer, models, trainers, pre_tokenizers, decoders, processors

corpus = '/kaggle/working/corpus.txt'
with open(corpus, 'w', encoding='utf-8') as f:
    for c in all_code:
        f.write(c + '\n')
    for conv in all_conversations:
        f.write(conv + '\n')

tok = Tokenizer(models.BPE())
tok.pre_tokenizer = pre_tokenizers.ByteLevel(add_prefix_space=False)
tok.decoder = decoders.ByteLevel()
tok.post_processor = processors.ByteLevel(trim_offsets=True)

SPECIAL = ['<|pad|>', '<|bos|>', '<|eos|>', '<|user|>', '<|assistant|>', '<|code_start|>', '<|code_end|>', '<|fim_prefix|>', '<|fim_middle|>', '<|fim_suffix|>']
trainer = trainers.BpeTrainer(vocab_size=8192, special_tokens=SPECIAL, min_frequency=1, show_progress=True)
tok.train([corpus], trainer)
tok.save('/kaggle/working/tokenizer.json')

VOCAB = tok.get_vocab_size()
PAD = tok.token_to_id('<|pad|>')
BOS = tok.token_to_id('<|bos|>')
EOS = tok.token_to_id('<|eos|>')
print(f'Vocab: {VOCAB}, PAD={PAD}, BOS={BOS}, EOS={EOS}')
```

## PHASE 3: INITIALIZE MODEL FROM SCRATCH

```python
config = SuperConfig(
    d_model=768, n_heads=12, n_kv_heads=4, n_layers=16, d_ff=2048,
    n_experts=4, n_active_experts=1, max_seq_len=2048,
    vocab_size=VOCAB, n_brain_heads=12, dropout=0.1,
    use_rope=True, use_flash_attn=True, rope_theta=1000000.0,
)

est = config.estimate_params()
print(f'Arch: d={config.d_model}, L={config.n_layers}, H={config.n_heads}, E={config.n_experts}')
print(f'Params: {est["total_billions"]:.3f}B ({est["total"]:,})')

model = SuperNeuralCore(config).to(DEVICE)
total = sum(p.numel() for p in model.parameters())
print(f'Actual: {total:,} ({total/1e9:.3f}B)')
print(f'Initialized from scratch on {DEVICE}')
```

## PHASE 4: TOKENIZE DATASET

```python
def tok_text(text):
    return tok.encode(text).ids[:config.max_seq_len]

print('Tokenizing...')
all_ids = [tok_text(c) for c in all_conversations if len(tok_text(c)) >= 10]
print(f'{len(all_ids)} tokenized seqs, avg {np.mean([len(t) for t in all_ids]):.1f} tokens')
random.shuffle(all_ids)
split = int(len(all_ids) * 0.95)
train_ids = all_ids[:split]
val_ids = all_ids[split:]
print(f'Train: {len(train_ids)}, Val: {len(val_ids)}')
```

## PHASE 5: TRAINING LOOP

```python
BS = 2
GA = 8
LR = 3e-4
EPOCHS = 3

opt = torch.optim.AdamW(model.parameters(), lr=LR, betas=(0.9, 0.95), weight_decay=0.1)

def collate(ts):
    ml = min(max(len(t) for t in ts), config.max_seq_len)
    ids = torch.full((len(ts), ml), PAD, dtype=torch.long, device=DEVICE)
    lbls = torch.full((len(ts), ml), -100, dtype=torch.long, device=DEVICE)
    for i, t in enumerate(ts):
        sl = min(len(t), ml)
        ids[i, :sl] = torch.tensor(t[:sl], dtype=torch.long)
        lbls[i, :sl] = torch.tensor(t[:sl], dtype=torch.long)
    return ids, lbls

def batch_gen(data, bs):
    idx = list(range(len(data)))
    random.shuffle(idx)
    for i in range(0, len(idx), bs):
        yield collate([data[j] for j in idx[i:i+bs]])

se = len(train_ids) // (BS * GA)
ts = se * EPOCHS
sched = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=ts, eta_min=1e-5)
print(f'Steps/epoch: {se}, Total: {ts}')
```

```python
log = []
gs = 0
best_val = float('inf')

for ep in range(1, EPOCHS + 1):
    print(f'\n{"="*60}\nEPOCH {ep}/{EPOCHS}\n{"="*60}')
    model.train()
    losses = []
    opt.zero_grad()
    
    for bidx, (inp, lbl) in enumerate(batch_gen(train_ids, BS)):
        out = model(inp)
        sl = out['logits'][..., :-1, :].contiguous()
        sll = lbl[..., 1:].contiguous()
        loss = F.cross_entropy(sl.view(-1, VOCAB), sll.view(-1), ignore_index=-100)
        (loss / GA).backward()
        
        if (bidx + 1) % GA == 0:
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()
            sched.step()
            opt.zero_grad()
            gs += 1
            with torch.no_grad():
                acc = (sl.argmax(-1) == sll).float().mean().item()
            losses.append(loss.item())
            if gs % 10 == 0:
                avg = np.mean(losses[-50:] if len(losses) >= 50 else losses)
                print(f'  S{gs:>4}/{ts} | E{ep} | l: {loss.item():.4f} | avg: {avg:.4f} | acc: {acc:.4f}')
            log.append({'step': gs, 'epoch': ep, 'loss': loss.item(), 'acc': acc})
    
    el = np.mean(losses)
    print(f'Epoch {ep}: train_loss={el:.4f}')
    
    model.eval()
    vls = []
    with torch.no_grad():
        for inp, lbl in batch_gen(val_ids, BS):
            v = F.cross_entropy(model(inp)['logits'][..., :-1, :].contiguous().view(-1, VOCAB), lbl[..., 1:].contiguous().view(-1), ignore_index=-100)
            vls.append(v.item())
    va = np.mean(vls)
    print(f'Val: loss={va:.4f} ppl={math.exp(min(va,20)):.2f}')
    
    ckpt = f'/kaggle/working/checkpoints/ep{ep}'
    os.makedirs(ckpt, exist_ok=True)
    torch.save({'model': model.state_dict(), 'epoch': ep, 'val': va}, f'{ckpt}/model.pt')
    if va < best_val:
        best_val = va
        torch.save({'model': model.state_dict(), 'config': config}, '/kaggle/working/best.pt')
        print(f'New best! val_loss={va:.4f}')
    gc.collect(); torch.cuda.empty_cache()

print(f'Done. Steps: {gs}, Best: {best_val:.4f}')
```

## RLCF: REINFORCEMENT FROM COMPILER FEEDBACK

```python
def compile_reward(code):
    try:
        ast.parse(code)
        return 1.0
    except SyntaxError:
        return -0.5

@torch.no_grad()
def generate(pid, mn=100, temp=0.8, tk=40):
    model.eval()
    inp = torch.tensor([pid], dtype=torch.long, device=DEVICE)
    for _ in range(mn):
        lo = model(inp)['logits'][:, -1, :] / temp
        v, _ = torch.topk(lo, tk, -1)
        lo[lo < v[:, -1:]] = float('-inf')
        nx = torch.multinomial(F.softmax(lo, -1), 1)
        inp = torch.cat([inp, nx], -1)
        if nx.item() == EOS:
            break
    return inp[0].tolist()

print(f'Valid code: {compile_reward("def f(): return 1")}')
print(f'Invalid: {compile_reward("def f(: return")}')
```

```python
cs_id = tok.encode('<|code_start|>').ids
ropt = torch.optim.AdamW(model.parameters(), lr=1e-5, weight_decay=0.01)
NRL = 50
rlog = []
for step in range(NRL):
    model.train()
    p = random.choice(train_ids)[:random.randint(20, 60)] + cs_id
    gen = generate(p, mn=60)
    txt = tok.decode(gen[len(p):]).split('<|code_end|>')[0].split('<|eos|>')[0]
    r = compile_reward(txt)
    inp = torch.tensor([gen], dtype=torch.long, device=DEVICE)
    out = model(inp)
    sh = out['logits'][0, :-1]
    tg = inp[0, 1:]
    loss = F.cross_entropy(sh.view(-1, VOCAB), tg.view(-1), ignore_index=PAD)
    ropt.zero_grad()
    (loss * max(0, 1.0 - r)).backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
    ropt.step()
    rlog.append({'r': r, 'l': loss.item()})
    if step % 10 == 0:
        print(f'RLCF {step}/{NRL} | r={r:.2f} | avg_r={np.mean([x["r"] for x in rlog[-20:]]):.2f}')
print(f'RLCF done. Final avg reward: {np.mean([x["r"] for x in rlog[-20:]]):.3f}')
```
